# POPULARITY BASED RECOMMONDER SYSTEM
(Top 50 books with highest average rating with minimum of 250 votes)

```python
books
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
##merging the three created data frames
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,5)) 
plt.hist(avg_rating_df['avg_rating'])
```

## COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_users=x[x].index
print(x[x].index)
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= y[y].index##boolean indexing
print(y[y].index)
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
similarity_scores=cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #fetch the index number
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        ##print(pt.index[i[0]])
        item = []
        temp_df = (books[books['Book-Title']== pt.index[i[0]]])
       # print(temp_df.drop_duplicates('Book-Title')['Book-Author'])
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
#recommend('Harry Potter and the Chamber of Secrets (Book 2)')
recommend('1984')
```
